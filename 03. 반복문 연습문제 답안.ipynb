# 연습문제 7
randint함수를 사용하여 정수(1~1000)로 이루어진 길이가 100개인 리스트 a를 만들고   
가장 큰 수를 구하는 코드를 작성하시오.  

예) a=[32, 45, 2, 5, 76] 일때, 출력결과는 76이 나와야 함.  





```python
import random
a = []
# 코드 작성

for i in range(100):
    a.append(random.randint(1,1000))

print("list:", a)

max_a = a[0]
for i in a:
    if max_a<i:
        max_a = i

print("가장 큰 수:", max_a)


# ## 또는
# max_a = max(a)
# print("가장 큰 수:", max_a)

# ## 또는
# # max_a = a.sort()
# a.sort(reverse=True) # 정렬후에
# max_a = a[0]
# print("정렬후 list: ", a)
# print("가장 큰 수:", max_a)

```

# 연습문제 8
randint함수를 사용하여 정수(1~1000)로 이루어진 길이가 100개인 리스트 a를 만들고 
리스트의 모든 요소의 합계를 구하는 코드를 작성하시오.

예) a=[32, 45, 2, 5, 76] 일때, 출력결과는 160이 나와야 함.







```python
import random
a = []
# 코드 작성

for i in range(100):
    a.append(random.randint(1,1000))

print("list:", a)

sum_a = 0
for i in a:
    sum_a += i

print("합계:", sum_a)

## 또는
# print("합계:", sum(a))
```

# 연습문제 9-1
word = [“school”, “game”, “piano”, “science”, “hotel”, “mountain”] 중   
글자수가 6글자 이상인 문자를 모아 새로운 리스트를 생성하세요  
**(리스트 컴프리핸션을 사용해주세요)**
